```python
data_set = pd.read_csv("DailyLevel_data_Lakewood.csv",error_bad_lines=False, engine="python")
```

```python
bus_list=data_set.business_name.unique()
```

```python
bus_list.sort()
```

```python
bus_list
```

```python
from gensim.parsing.preprocessing import remove_stopwords
new_list = []
for i in bus_list:
  i = remove_stopwords(i)
  new_list.append(i) 
len(new_list)
```

```python
%matplotlib inline
plt.style.use('bmh')
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
# connect to google
pytrends = TrendReq(hl='en-US', tz=360)
import numpy as np
```

```python
d = { 'Google interest': []}
Complete_df = pd.DataFrame(data=d)
for bus_name in new_list:
  # keywords to search for
  search = pd.DataFrame()
  sum1 = pd.DataFrame(data=d)
  ls= bus_name.split() 
  pytrends.build_payload(kw_list=ls,timeframe="2017-12-31 2018-10-31",geo='US-NJ')
  # dataframe
  search = pytrends.interest_over_time()
  # print((search))
  first = True
  for index, row in search.iterrows():
    np_row = np.array(row)
    n = 6 if first else 4 if index == search.index[-1] else 7
    first = False
    for i in range(n):
        summing = int(np.sum(np_row[0:-1])/len(np_row))
        sum1.loc[len(sum1.index)] = [summing]
  # print(Complete_df)
  Complete_df = Complete_df.append(sum1, ignore_index = False)
#Plot a graph of google interest vs check ins for all together 



# df.to_csv('gtrends11.csv')
# files.download("gtrends11.csv")
# creating graph
# fig, ax = plt.subplots(figsize=(12, 6))
# df.head(7)
# df['Big'].plot(color='purple')
# adding title and labels
# plt.title('Total Google Searches for '+ls[0], fontweight='bold')
# plt.xlabel('Year')
# plt.ylabel('Total Count')
```

```python
Complete_df.size
Complete_df = Complete_df.reset_index(drop=True)
```

```python
sorted_data = data_set.sort_values(by=['business_name'])
```

```python
concat = pd.concat([sorted_data, Complete_df], axis=1)
```
