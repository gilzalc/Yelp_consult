## Data transforming

### categories 

```python
df = pd.read_csv('/content/business_details.csv')
# df.head()
```

```python
categories = ['American','Italian','Asia', 'breakfast','bar','others']
for i in categories:
    df[i]=0
```

```python
df['categories'] = df['categories'].apply(lambda x: x.lower())
```

```python
American_list = ['american','steakhouse','chicken','burger','hamburger','steak','fast Food','donuts']
Asian_list = ['dim sum','sushi','japanese','chinese','thai','viet','vietnamese','taiwanese','noodles','ramen','curry','malaysian','cambodian']
Bar_list = ['bar','pub','cocktails','drinks','party','club']
breakfast_list =  ['brunch','coffee','bagel','breakfast','tea','sandwich','bakeries','delis']
italian_list = ['italian','pizza','pasta','gelato']
```

```python
for index, category in df['categories'].iteritems():
  flag=False
  for x in American_list:
    if x in category:
      flag = True
      df.at[index,'American'] = 1
      break
  for x in Asian_list:
    if x in category:
      flag = True
      df.at[index,'Asia'] = 1
      break
  for x in Bar_list:
    if x in category:
      flag = True
      df.at[index,'bar'] = 1
      break
  for x in breakfast_list:
    if x in category:
      flag = True
      df.at[index,'breakfast'] = 1
      break
  for x in italian_list:
    if x in category:
        flag = True
        df.at[index,'Italian'] = 1
        break
  if flag == False:
    df.at[index,'others'] = 1
```

```python
# Get data about the opening hours of each bussiness
for index, id in df['id'].iteritems():
    if index < 500:
          continue
    print(index,id)
    url = f"https://api.yelp.com/v3/businesses/{id}"
    api_key = "QXl5o-ixKcN5K3XVUhqUWQVALpRLFrDtpB7CSqPBFi7V_oeyh9VmFhpz8sM6fnEnkTt8LsXbBjVj2V-Ugh1udzKuswhusf3mjEGF6L-gN4zmrfBWE8QX8pEkAVdmZHYx"
    # Make a request to the Yelp API
    headers = {"Authorization": f"Bearer {api_key}"}
    response = requests.get(url, headers=headers)
    data = response.json()
    print(json.dumps(data, indent=4))
    if ('error' in data):
      break
    if(len(data.get("hours", []))==0):
      print ("**********")
      df.at[index, 'is_overnight'] = None
      df.at[index,'openWeekend'] =  None
      continue
    for day in data.get("hours", [])[0]['open']:
      if day['is_overnight']:
        df.at[index, 'is_overnight'] = True
      if day['day'] in [5,6]:
        df.at[index,'openWeekend'] =  True
```
